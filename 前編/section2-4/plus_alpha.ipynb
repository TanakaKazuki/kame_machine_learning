## ax.contour の仕様
- 指定された X と Y の格子点上で計算された値 Z を用いて等高線図を描画する関数
    - X, Y：これらは np.meshgrid によって生成された二次元配列で、関数の各点の x と y 座標を表します。
    - Z：関数の各点での値を含む二次元配列で、X と Y の各点に対応します。
    - levels：描画する等高線のレベルを指定するリストです。このリストの値は、Z において特定の値を持つ点を結ぶ等高線が描かれます。

```python
import numpy as np
import matplotlib.pyplot as plt

# z = 2(x + y)
def f(x, y):
    return 2 * (x + y)

x = [-1, 0, 1, 2, 3]
y = [-1, 0, 1, 2, 3]

X, Y = np.meshgrid(x, y)

fig, ax = plt.subplots()
cont = ax.contour(X, Y, f(X, Y), levels=[-4, -2, 0, 2, 4]) # ここだけ説明。
cont.clabel(fmt='%1.1f', fontsize=14)
plt.show()
```
