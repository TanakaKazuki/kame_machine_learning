## 損失関数

```python
def cost_func(theta0, theta1, x, y):
    sigma = np.square(y - (theta0 + theta1 * x))
    loss = np.mean(sigma)
    return loss

x = df["space"].values
y = df["rent"].values

cost_func(2,1,x,y)
```

### 最急降下法によるパラメータ更新

#### 初期値の設定

```python
# 指定ver
theta_0_init = -5
theta_1_init = -0.5

# ランダムver
theta_0_init = np.random.uniform(min_theta0, max_theta0)
theta_1_init = np.random.uniform(min_theta1, max_theta1)

print(theta_0_init,theta_1_init)
```

#### パラメータ更新の関数定義

```python
def update_theta0(theta0, theta1, x, y, alpha=0.00005):
    theta0 = theta0 - alpha * 2 * np.mean((theta0 + theta1 * x) - y)
    return theta0

def update_theta1(theta0, theta1, x, y, alpha=0.00005):
    theta1 = theta1 - alpha * 2 * np.mean(((theta0 + theta1 * x) - y) * x)
    return theta1
```

```python
theta_0_init
```

```python
update_theta0(theta_0_init, theta_1_init, x, y)
```

```python
theta_1_init
```

```python
update_theta1(theta_0_init, theta_1_init, x, y)
```

#### イテレーションによるepoch管理

```python
epochs = 100000
alpha = 0.00005

theta_0_hist = []
theta_1_hist = []
theta_0_hist.append(theta_0_init)
theta_1_hist.append(theta_1_init)
for _ in range(epochs):
    updated_theta_0 = update_theta0(theta_0_hist[-1],theta_1_hist[-1],x,y,alpha)
    updated_theta_1 = update_theta1(theta_0_hist[-1],theta_1_hist[-1],x,y,alpha)
    
    theta_0_hist.append(updated_theta_0)
    theta_1_hist.append(updated_theta_1)
```

```python
# 最終的なパラメータ
print(theta_0_hist[-1])
print(theta_1_hist[-1])
```

#### 損失関数の推移を3D・2D（等高線）で計算
#### 得られたthetaで回帰直線を描画

```python
fig = plt.figure(figsize=(12, 6))
ax = plt.axes(projection = '3d')
ax.plot_surface(theta_0_ax, theta_1_ax, Z, cmap='jet', alpha=0.5)
ax.plot(theta_0_hist, theta_1_hist, [cost_func(t0, t1, x, y) for t0, t1 in zip(theta_0_hist, theta_1_hist)], 'r.-') # 更新履歴
ax.view_init(elev=20, azim=45)
```

```python
# 2D等高線プロット
fig = plt.figure(figsize=(12, 6))
ax2 = fig.add_subplot(122)
contour = ax2.contour(theta_0_ax, theta_1_ax, Z, levels=50, cmap='jet')
fig.colorbar(contour, ax=ax2, orientation='vertical', label='Loss Magnitude')
ax2.set_xlabel('$\\theta_0$')
ax2.set_ylabel('$\\theta_1$')
ax2.set_title('Cost Function Contour')

# パラメータ更新履歴のプロット
ax2.plot(theta_0_hist, theta_1_hist, 'r.-') # 更新履歴を赤点で表示
ax2.plot(theta_0_hist[-1], theta_1_hist[-1], 'ro')  # 最終的なパラメータ位置を赤い丸で表示

# 回帰直線のプロット
# xデータと予測yデータ（最新のthetaを使用）
y_pred = theta_0_hist[-1] + theta_1_hist[-1] * x
ax3 = fig.add_subplot(223)
ax3.scatter(x, y, color='blue', label='Actual data')  # 実データの散布図
ax3.plot(x, y_pred, color='red', label='Fitted line')  # 回帰直線
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_title('Regression Line')
ax3.legend()

# 全体図の調整
plt.tight_layout()
plt.show()
```

### 最急降下法の注意点
- 最適解ではなく、局所解に辿り着くこともある
- 学習率が大きすぎるあまり、発散して学習が進まない
- いろんなアルファを試すのもDSの仕事と言える

## 正規方程式


```python
# 特徴量行列で1列目は要素がすべて1
np.ones(len(x))
# これを列方向にxに加える
np.vstack([np.ones(len(x)),x]).T
```

```python
X = np.vstack([np.ones(len(x)),x]).T
```

```python
# 正規方程式の再現
best_theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
best_theta
```
