## t検定

```python
import pandas as pd
df = pd.read_csv('../section5/sample_data2.csv')
```

```python
df
```

```python
#目的変数
y_col = 'rent'
X = df.drop(columns=[y_col])
```

```python
X
```

```python
# 線形回帰
import statsmodels.api as sma
#バイアス項（シータゼロの係数）の追加
X = sma.add_constant(X)
display(X)
y = df[y_col]
est = sma.OLS(y,X)
est_trained = est.fit()
est_trained.summary()
```

```python
# 標準化してみる
from sklearn.preprocessing import StandardScaler
X = df.drop(columns=[y_col])
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
X_scaled = sma.add_constant(X_scaled)
```

```python
X_scaled
```

```python
est = sma.OLS(y,X_scaled)
est_trained = est.fit()
est_trained.summary()
```
